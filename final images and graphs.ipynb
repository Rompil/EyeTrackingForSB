```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.image as mpimg
%matplotlib inline

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('.\\final_best_result_for_eyeTracking.dat', sep = ' ', names = ['lx', 'ly', 'lr', 'rx', 'ry', 'rr'] )
```

```python
plt.plot( 'lx', 'ly', data=df, linestyle=':', marker='o')
#plt.show()
plt.savefig('left.eye.png')
```

```python
left_eye = []
right_eye = []
with open('.\\final_best_result_for_eyeTracking.dat','r') as f:
    for line in f:
        t = line.split()
        t = [float(f) for f in t]
        l, r = tuple(t[:2]),tuple(t[3:5])
        left_eye.append(l)
        right_eye.append(r)
```

```python
fig = plt.figure()
fig.set_size_inches(18.5, 15.5, forward=True)
ax1 = fig.add_subplot(211)
ax1.plot(*zip(*left_eye))

ax2 = fig.add_subplot(212)
ax2.plot(*zip(*right_eye))
plt.savefig('better left and right eyes.png')
plt.show()
```
