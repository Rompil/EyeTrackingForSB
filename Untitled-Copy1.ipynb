```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
left_eye = []
right_eye = []
with open('test.dat','r') as f:
    for line in f:
        t = ''.join([c for c in line if c not in '(),']).split()
        t = [float(f) for f in t]
        l, r = tuple(t[:2]),tuple(t[2:])
        left_eye.append(l)
        right_eye.append(r)
```

```python
x, y = map(list, zip(*left_eye))
```

```python

fig = plt.figure()
fig.set_size_inches(18.5, 15.5, forward=True)
ax1 = fig.add_subplot(211)
ax1.plot(*zip(*left_eye))

ax2 = fig.add_subplot(212)
ax2.plot(*zip(*right_eye))

plt.show()
```
