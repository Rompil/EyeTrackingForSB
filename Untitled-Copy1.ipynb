```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
image = mpimg.imread(r'.\\data\\leftEye0.jpg ')
print(image.shape)
Bimage = cv2.edgePreservingFilter(image,flags=1, sigma_s=10)
imgplot = plt.imshow(Bimage)
```

```python
image = cv2.cvtColor(Bimage, cv2.COLOR_RGB2GRAY)
```

```python
imgplot = plt.imshow(image)
```

```python
image.shape

neg_image = np.subtract(np.full(image.shape, 255), image).astype('uint8')


imgplot = plt.imshow(neg_image)
```

```python
import  cv2

Gimage = cv2.GaussianBlur(image, (5,5), 2, 0)
Mimage = cv2.medianBlur(image, 5)
#Gimage = cv2.medianBlur(Gimage, 3)

imgplot = plt.imshow(Gimage)
```

```python
imgplot = plt.imshow(Mimage)
```

```python
Center_min =np.argwhere(Gimage == np.min(Gimage))
#print(Center_min)
```

```python
c_g_neg_img = cv2.Canny(Mimage, 1, 20, 10, 3 )

imgplot = plt.imshow(c_g_neg_img)
circles = cv2.HoughCircles(Mimage, cv2.HOUGH_GRADIENT, 4, 1, param1 = 20, param2 = 10, minRadius = 0, maxRadius = 40)
print(circles)
```

```python
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(c_g_neg_img,(i[0],i[1]),i[2],(100,100,70),2)
    # draw the center of the circle
    cv2.circle(c_g_neg_img,(i[0],i[1]),3,(30,30,30),3)
    print(i[0],i[1])
imgplot = plt.imshow(np.vstack([image, Mimage, c_g_neg_img]))
```

```python
a_c = np.mean(Center_min, axis=0)
for c in Center_min:
    cv2.circle(image,(c[1],c[0]),0,(255,255,0),2)
    pass
cv2.circle(image,(int(a_c[1]),int(a_c[0])),0,(255,0,100),2)
implot = plt.imshow(image)   
```
